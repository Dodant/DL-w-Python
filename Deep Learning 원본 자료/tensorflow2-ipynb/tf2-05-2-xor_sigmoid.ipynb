```
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
```

```
def sigmoid(x):
    return(1. / (1. + np.exp(-x)))
```

```
x_data = np.array([[0., 0.],
          [0., 1.],
          [1., 0.],
          [1., 1.]])
y_data = np.array([[0.],
          [1.],
          [1.],
          [0.]])

W1 = np.array([[5., 5.], [-7.,-7.]])
b1 = np.array([[-8., 3.]])
W2 = np.array([[-11., -11.]])
b2 = np.array([[6.]])

a1 = np.add(np.matmul(x_data, np.transpose(W1)), b1)
z1 = sigmoid(a1)
a2 = np.add(np.matmul(z1, np.transpose(W2)), b2)
y = sigmoid(a2)

y_pred = np.array(y>0.5, dtype=np.int32)
```

```
print(a1)
print(z1)
print(a2)
print(y)
print(y_pred)
```
